```python
def standardize_name(clean_fullname):
    clean_fullname = clean_fullname.strip()
    if clean_fullname == "SEAN PARKER AND AFFILIATED ENTITIES":
        return "SEAN PARKER"
    elif clean_fullname == "FUND FOR POLICY REFORM (NONPROFIT 501 (C)(4))":
        return "FUND FOR POLICY REFORM"
    else:
        return clean_fullname
```

```python
merged["standardized_name"] = merged.clean_fullname.apply(standardize_name)
```

```python
bigdogs = merged.groupby(
    ["standardized_name", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
